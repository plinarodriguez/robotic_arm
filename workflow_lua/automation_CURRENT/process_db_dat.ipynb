```python
import pandas as pd

# ----------------------------
# Input: rows 6–16 as provided
# ----------------------------
ROWS = {
    
    1:  dict(start_x = 136, start_y = -296.5,    dx_steps = [57.5, 55.5, 59.0, 56.5, 58.0], dy_steps = [0.5, 1.0, 0.5, 2.0, 1.0], bottles = 6 ),
    2:  dict(start_x = -122.5, start_y = -309.0, dx_steps = [57.5, 57.5, 57.5, 56.5      ], dy_steps = [1.0, 1.0, 2.0, 0.5]     , bottles = 5 ),
    3:  dict(start_x = 135.5, start_y = -330.5,  dx_steps = [57.5, 57.5, 57.0, 58.0, 57.5], dy_steps = [0.5, 1.5, 1.0, 1.5, 2.0], bottles = 6 ),
    4:  dict(start_x = -123.5, start_y = -341.0, dx_steps = [58.0, 56.0, 58.5, 57.0      ], dy_steps = [1.5, 2.5, 0.5, 1.5]     , bottles = 5 ),
    5:  dict(start_x = 134.5, start_y = -363.0,  dx_steps = [56.5, 58.0, 57.5, 58.0, 56.5], dy_steps = [0.5, 1.0, 1.0, 1.5, 2.0], bottles = 6 ),  
    6:  dict(start_x=-123.5, start_y=-374.5, dx_steps=[57.5, 57.5, 57.5, 58.0],      dy_steps=[2.0, 1.5, 0.5, 1.5],   bottles=5),
    7:  dict(start_x=134.5,  start_y=-397.0, dx_steps=[56.5, 58.0, 57.5, 58.0, 57.0], dy_steps=[0.5, 1.0, 2.5, 1.0, 2.0], bottles=6),
    8:  dict(start_x=-124.0, start_y=-407.0, dx_steps=[57.0, 57.5, 58.5, 57.5],      dy_steps=[2.0, 1.5, 1.5, 1.5],   bottles=5),
    9:  dict(start_x=134.0,  start_y=-430.0, dx_steps=[56.5, 58.0, 58.0, 58.0, 57.0], dy_steps=[0.5, 1.0, 2.5, 1.0, 2.0], bottles=6),
    10: dict(start_x=-125.0, start_y=-440.0, dx_steps=[57.0, 57.5, 58.5, 57.5],      dy_steps=[2.0, 1.5, 1.5, 0.5],   bottles=5),
    11: dict(start_x=133.0,  start_y=-463.0, dx_steps=[56.5, 58.0, 58.0, 57.5, 57.0], dy_steps=[1.5, 2.0, 0.5, 1.0, 2.0], bottles=6),
    12: dict(start_x=-125.5, start_y=-474.0, dx_steps=[57.0, 57.5, 58.5, 57.5],      dy_steps=[2.0, 0.0, 1.5, 1.5],   bottles=5),
    13: dict(start_x=132.5,  start_y=-496.5, dx_steps=[56.5, 58.0, 58.0, 57.5, 57.0], dy_steps=[1.5, 2.0, 0.5, 1.0, 2.0], bottles=6),
    14: dict(start_x=-126.0, start_y=-507.5, dx_steps=[57.5, 57.5, 57.5, 57.5],      dy_steps=[0.5, 1.5, 1.0, 1.5],   bottles=5),
    15: dict(start_x=131.5,  start_y=-528.5, dx_steps=[56.5, 58.0, 57.5, 57.5, 57.0], dy_steps=[1.5, 1.0, 0.5, 1.0, 2.0], bottles=6),
    16: dict(start_x=-126.5, start_y=-540.5, dx_steps=[55.0, -4.0, 56.47, 119.03],     dy_steps=[1.5, 24.5, 1.75, -24.25], bottles=5),
}

def compute_xy_positions(rows: dict):
    """
    Reproduce Lua xy_for() logic:
      odd rows:  x -= dx, y += dy
      even rows: x += dx, y -= dy

    Returns:
      df_x: X positions per bottle index (0..max_bottles-1)
      df_y: Y positions per bottle index (0..max_bottles-1)
      df_xy: long-format (row, col, x, y) with col being 1-based like Lua
    """
    max_bottles = max(v["bottles"] for v in rows.values())
    cols = list(range(max_bottles))  # bottle index within row (0 = bottle 1)

    df_x = pd.DataFrame(index=sorted(rows.keys()), columns=cols, dtype=float)
    df_y = pd.DataFrame(index=sorted(rows.keys()), columns=cols, dtype=float)

    for r in sorted(rows.keys()):
        row = rows[r]
        x = row["start_x"]
        y = row["start_y"]
        df_x.loc[r, 0] = x
        df_y.loc[r, 0] = y

        odd = (r % 2 == 1)  # same convention as your Lua code

        for i in range(1, row["bottles"]):
            dx = row["dx_steps"][i-1]
            dy = row["dy_steps"][i-1]
            if odd:
                x = x - dx
                y = y + dy
            else:
                x = x + dx
                y = y - dy
            df_x.loc[r, i] = x
            df_y.loc[r, i] = y

    # Long format is great for debugging / comparing to taught points
    records = []
    for r in df_x.index:
        for c0 in df_x.columns:
            x = df_x.loc[r, c0]
            y = df_y.loc[r, c0]
            if pd.notna(x) and pd.notna(y):
                records.append({"row": int(r), "col": int(c0 + 1), "x": float(x), "y": float(y)})
    df_xy = pd.DataFrame(records)

    return df_x, df_y, df_xy

df_x, df_y, df_xy = compute_xy_positions(ROWS)

print("X table:\n", df_x)
print("\nY table:\n", df_y)
print("\nLong table:\n", df_xy)

# Optional: save CSVs
df_x.to_csv("tray_x_positions_rows6_16.csv", index_label="row")
df_y.to_csv("tray_y_positions_rows6_16.csv", index_label="row")
df_xy.to_csv("tray_xy_positions_rows6_16_long.csv", index=False)
```
